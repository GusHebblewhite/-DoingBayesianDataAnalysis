# Ch 7_2 - MCMC Multiple Parameters
Say we have 2 independant\* coins with p(heads) = $\theta_j$ (where j $\in$ {1,2}) which are flipped $H_j$ times and result in $z_j$ heads. The $i^{th}$ individual flip of coin $j$ is denoted $y_{ji}$.

\*Idependence is a very important assumption. In general, experiments will be designed to ensure that this is the case.

Note though that there are different ways in which a model might (or might not) be independent. For example;
1. The parameters of each of the coins can be independent of the others. If this is true, it means that $p(\theta_1, \theta_2) = p(\theta_1)p(\theta_2)$.
2. The coin flips themselves be be independent of each other, and independent of the flips of the other coin(s). This is usually true by experimental design. If this is true, it means that $p(y_j \, | \, \theta_1, \theta_2) = p(y_j \, | \, \theta_j)$ where $j \in \{1,2\}$.

##### Priors:
Our prior beliefs are about the *combinations* of paramter values, and so with 2 coins, the density function '$p(\theta_1, \theta_2)$' of our priors will be two dimensional and must still integrate to 1, i.e.
$$\int \int p(\theta_1, \theta_2) \; d\theta_1 \, d\theta_2 = 1$$

$D_j$ = the observations of the flips of coin j  
$z_j = \sum_{i=1}^{N_j} y_{ij}$

##### Likelihoods:
**Importantly** because the coin flips are independent, $p(y_j \, | \, \theta_1, \theta_2) = p(y_j \, | \, \theta_j)$. Hence:

$$p(D \, | \, \theta_1, \theta_2) = \prod_{y_{1i} \in D_1} p(y_{1i} \, | \, \theta_1, \theta_2) \times \prod_{y_{2j} \in D_2} p(y_{2j} \, | \, \theta_1, \theta_2)$$
$$= \prod_{y_{1i} \in D_1} \theta_1^{y_{1i}} (1-\theta_1)^{(1 - y_{1i})} \times \prod_{y_{2j} \in D_2} \theta_2^{y_{2j}} (1-\theta_2)^{(1 - y_{2j})}$$
$$= \theta_1^{z_1} (1-\theta_2)^{(N_1 - z_1} \; \theta_2^{z_2} (1-\theta_2)^{(N_2 - z_2)}$$

##### Posteriors:
$$p(\theta_1, \theta_2 \, | \, D) = \frac{p(D \, | \, \theta_1, \theta_2)\; p(\theta_1, \theta_2)}{p(D)}$$

$$p(D) = \int \int p(D \, | \, \theta_1, \theta_2)\; p(\theta_1, \theta_2) \; d\theta_1 \, d\theta_2$$

### Analytic example:
Lets use the Beta Distribution for our prior. Assuming independence assumption 1) is met, both $\theta_1$ and $\theta_2$ will be independent beta distributions. Using this prior, our posterior becomes:
$$p(\theta_1, \theta_2 \, | \, D) = \frac{\theta_1^{z_1} (1-\theta_2)^{(N_1 - z_1)} \; \theta_2^{z_2} (1-\theta_2)^{(N_2 - z_2)} \; \textrm{beta}(\theta_1, a_1, b_1) \; \textrm{beta}(\theta_2, a_2, b_2)}{p(D)}$$

$$ = \frac{\theta_1^{z_1} (1-\theta_2)^{(N_1 - z_1)} \; \theta_2^{z_2} (1-\theta_2)^{(N_2 - z_2)} \; \theta_1^{(a_1 - 1)} (1-\theta_2)^{(b_1 - 1)} \; \theta_2^{(a_2 - 1)} (1-\theta_2)^{(b_2 - 1)}} {B(a_1, b_1) \; B((a_2, b_2) \; p(D)}$$

$$ = \frac{\theta_1^{(z_1 + a_1 - 1)} (1-\theta_2)^{(N-1 - z_1 + b_1 - 1)} \; \theta_2^{(z_2 + a_2 - 1)} (1-\theta_2)^{(N_2 - z_2 + b_2 - 1)}} {B(a_1, b_1) \; B((a_2, b_2) \; p(D)}$$

Since the denominator is merely a normalizing term, and the top has been simplified to the product of two beta distributions, we can also simplify the denominator to:
$$B(z_1 + a_1, N_1 - z_1 + b_1) \; B(z_2 + a_2, N_2 - z_2 + b_2)$$

### Monte Carlo Example (Metropolis)
Since we are now able to express our posterior for the bivariate case, we can now try to sample using the metropolis algorithm. Since we are choosing two paramters, we can use the bivariate normal (a round normal-shaped mound) to generate points. 

```r
#2 dimensional MCMC (metropolis)

#Prior:
a_1 = 2
b_1 = 2
a_2 = 2
b_2 = 2

get_prior = function(theta_1, a_1, b_1, theta_2, a_2, b_2){
    return(dbeta(theta_1, a_1, b_1) * dbeta(theta_2, a_2, b_2))
}

#Evidence:
N_1 = 200
z_1 = 10
N_2 = 200
z_2 = 10

get_likelihood = function(a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2){
    return(beta(z_1 + a_1, N_1 - z_1 + b_1) * beta(z_2 + a_2, N_2 - z_2 + b_2))
}

#Normalising term (for beta only)
get_norm = function(theta_1, a_1, b_1, N_1, z_1, theta_2, a_2, b_2, N_2, z_2){
    return(beta(z_1 + a_1,  N_1 - z_1 + b_1) * beta(z_2 + a_2,  N_2 - z_2 + b_2)
          )
}

#Posterior:

#################################
# Beta
#################################

get_beta_posterior = function(theta_1, a_1, b_1, N_1, z_1, 
                              theta_2, a_2, b_2, N_2, z_2){
    
    #The posterior function for beta distributions
    posterior = (get_likelihood(a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2) * 
                 get_prior(theta_1, a_1, b_1, theta_2, a_2, b_2) /
                 get_norm(theta_1, a_1, b_1, N_1, z_1, theta_2, a_2, b_2, N_2, z_2)
                 )
}

get_beta_result = function(no_samples_per_dim, 
                              a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2){
    
    #generate possible theta values:
    x = seq(0, 1, 1/no_samples_per_dim)
    
    #get the distribution
    result = c()
    
    for (x1 in x){
        for (x2 in x){
            point = get_beta_posterior(x1, a_1, b_1, N_1, z_1, 
                                       x2, a_2, b_2, N_2, z_2)
            result = append(result, point)
        }
    }
    
    return(matrix(result, length(x), length(x)))
}

#################################
# MCMC metropolis
#################################

get_MCMC_posterior = function(theta_1, a_1, b_1, N_1, z_1, theta_2, a_2, b_2, N_2, z_2){
    
    point = (get_likelihood(a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2) * 
             get_prior(theta_1, a_1, b_1, theta_2, a_2, b_2))
    
    return(point)
    
}

get_MCMC_result = function(iter, 
                           a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2){
    
    #Randomly initialise the thetas:
    theta_1_curr = runif(1, 0, 1)
    theta_2_curr = runif(1, 0, 1)
    
    #Save
    result_1 = c(theta_1_curr)
    result_2 = c(theta_2_curr)
    
    #iterate
    for (i in seq(iter)){
        
        #Generate random prospective thetas
        theta_1_pros = rnorm(1, theta_1_curr, 0.2)
        theta_2_pros = rnorm(1, theta_2_curr, 0.2)
        
        #Metropolis
        p_curr = get_MCMC_posterior(theta_1_curr, a_1, b_1, N_1, z_1, theta_2_curr, a_2, b_2, N_2, z_2)
        p_pros = get_MCMC_posterior(theta_1_pros, a_1, b_1, N_1, z_1, theta_2_pros, a_2, b_2, N_2, z_2)
        
        if (runif(1, 0 ,1) < min(1, p_pros / p_curr)){
            theta_1_curr = theta_1_pros
            theta_2_curr = theta_2_pros
        }
        result_1 = append(result_1, theta_1_curr)
        result_2 = append(result_2, theta_2_curr)
    }
    
    return(table(cut(result_1, 100), cut(result_2, 100)))
}

num_iter = 100000
result_MCMC = get_MCMC_result(num_iter, a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2)
dim_size = 100
result_beta = get_beta_result(dim_size, a_1, b_1, N_1, z_1, a_2, b_2, N_2, z_2)
```

```r
#Plot:
library(plot3D)
library(repr)
options(repr.plot.width=8, repr.plot.height=4)

x = seq(0, 1, 1 / dim_size)

par(mfrow=c(1,2))
hist3D(z = result_MCMC)

scatter3D(z = result_beta, 
        x = rep(x, each = length(x)), 
        y = rep(x, times = length(x))
        )
```
